## Introduccion teorica

Los metodos de Tension Bien Medida (TBM) y Corriente Bien Medida (CBM) son utiles para medir el valor de una resistencia en un punto especifico, es decir, a una corriente y tension dadas. Las configuraciones circuitales de cada metodo son las siguientes:

<img src="./img/TBM_CBM.png"  width=550 /> 

Para ambos metodos la resistencia medida $R_I$ resulta $\frac{V_I}{I_I}$, sin embargo debido a la topologia de cada circuito, se debe aplicar una correcion diferente para obtener la $R_x$ real. Siendo $R_V$ la resistencia interna del voltimetro y $R_A$ la del amperimetro:

#### TBM
* $R_I = R_x // R_V$ por ende con este metodo se mide resistencia con defecto.
* $e_{MRT} = \frac{-R_I}{R_x}$
<img src="./img/Em_v.png"  width=350 />

#### CBM
* $R_I = R_x + R_A$ por ende con este metodo se mide resistencia con exceso.
* $e_{MRC} = \frac{R_A}{R_I-R_A}$
<img src="./img/Em_c.png"  width=350 />

Si se grafican los modulos de los errores de metodo en el mismo grafico, se obtiene lo siguiente:

<img src="./img/Em.png"  width=550 />

Donde la resistencia de decision corresponde a: $R_D = \sqrt{R_A R_V}$

## Calculos

```python
V_10_t = [1, 1, 1]
I_10_t = [0.100990, 0.100970, 0.100950]

Va_10_t = np.average(V_10_t)
Ia_10_t = np.average(I_10_t)
R_10_t = Va_10_t/Ia_10_t
print("R_10_t = ", R_10_t)

ua_V_10_t = ( 1/m.sqrt(n) * np.std(V_10_t) )
ua_I_10_t = ( 1/m.sqrt(n) * np.std(I_10_t) )

ubr_V_10_t = ( 1/m.sqrt(3) * Va_10_t * (erv/100 + n_er/1000) )
ubr_I_10_t = ( 1/m.sqrt(3) * Ia_10_t * (eri/100 + n_er/10097) )

ur_V_10_t = m.sqrt(ua_V_10_t**2 + ubr_V_10_t**2) / Va_10_t
ur_I_10_t = m.sqrt(ua_I_10_t**2 + ubr_I_10_t**2) / Ia_10_t

U_R_10_t = m.sqrt(ur_V_10_t**2 + ur_I_10_t**2) * R_10_t * k  
print("U_R_10_t = ", U_R_10_t)
```
