## Objetivo
Calibrar, en tensión y en corriente, un multímetro genérico. Se entiende por calibración, al traspaso de la incertidumbre de un patrón al instrumento a calibrar.  
Para ello se proponen varios puntos de calibración que deben ser usados como referencia en todas las escalas del instrumento posibles.
Una vez obtenidos los valores y su incertidumbre, evaluar la veracidad de la hoja de datos del equipo calibrado contrastando los valores obtenidos frente al semi rango definido por el fabricante.

Lo circuitos a evaluar son los siguientes:

Donde $V_{i}$ es el valor indicado por el voltímetro genérico y $V_{p}$ es el indicado por el voltímetro patrón.

Donde $A_{i}$ es el valor indicado por el amperímetro genérico y $A_{p}$ es el indicado por el amperímetro patrón.


## Desarrollo

Para determinar la incertidumbre y la desviación en cada uno de los puntos de calibración utilizados como referencia se utilizará respectivamente:

$S_{i} = (\frac{e\%}{100} + \frac{2}{V_{m} \cdot f_{c}}) \cdot V_{m} \cdot \frac{k}{\sqrt3}$

$Δ_{i} = V_{m} - V_{p}$
